# Segmentação do sinal em frames de 20ms

## Normalizando o sinal

```python
sinal = sinal / max(abs(sinal))                                                                        
t=(np.arange(len(sinal)))/fs
assert min(sinal) >= -1 and max(sinal) <= 1
```

```python
plt.figure(figsize=(20,5))
plt.figure(1), plt.title('Sinal Original'), plt.grid()
plt.plot(sinal), plt.xlabel('Amostras'), plt.ylabel('Amplitude Normalizada'),plt.grid 
```

```python
amostras_ms = int(fs / 1000)
frame_time = 20
amostras_frame = amostras_ms * frame_time
nFrames = int(len(sinal) / amostras_frame)      
print(f'Número de amostras/frames : {amostras_frame}')
print ('Número de frames:', nFrames)
```

# Calcule e plote a energia segmental

\begin{equation}
E_n = \sum_{m=-\inf}^{\inf}[x^2(m)*h(n-m)]
\end{equation}

\begin{equation}
h(n) = w^2(n)
\end{equation}

```python
STEs = []                                      

for k in range(nFrames):
    startIdx = k * amostras_frame
    stopIdx = startIdx + amostras_frame
    window = np.zeros(sinal.shape)
    window[startIdx:stopIdx] = 1              
    STE = sum((sinal ** 2) * (window ** 2))
    STEdB = 10*np.log10(STE)                  
    STEs.append(STEdB)

plt.figure(figsize=(10,5))    
plt.grid()
plt.plot(STEs)
plt.title('Energia Segmental')
plt.ylabel('Energia(dB)')
plt.xlabel('Frame')
plt.show()
```

# Calcule e plote a taxa de cruzamento por zeros (TCZ) segmental

\begin{equation}
	TCZ_n = \sum_{m=-\inf}^{\inf}|sgn[x(m)] - sgn[x(m-1)]|w(n-m)
\end{equation}


\begin{equation}
	sgn[x(n)] = 
	\begin{cases}
		1, & {x(n)} \geq 0 \\
		-1,& x(n) < 0
	\end{cases}
\end{equation}


\begin{equation}
w(n) = 
	\begin{cases}
		\frac{1}{2N}, & 0\leq n \leq N-1 \\
		0, & \mbox{c.c}
	\end{cases}
\end{equation}

```python
DC = np.mean(sinal)
novoSinal = sinal - DC               
```

```python
amostras_frame
```

```python
ZCCs = []                                     
for i in range(nFrames):
    startIdx = i * amostras_frame
    stopIdx = startIdx + amostras_frame
    s = novoSinal[startIdx:stopIdx]           
    ZCC = 0
    #Implementação da Equação
    for k in range(1, len(s)):
        ZCC += 0.5 * abs(np.sign(s[k]) - np.sign(s[k - 1]))
    ZCCs.append(ZCC)
```

```python
plt.figure(figsize=(10,6)) 
plt.plot(ZCCs)
plt.grid()
plt.title('TCZ'),plt.ylabel('TCZ'), plt.xlabel('Frame')
```

# Implemente uma rotina que identifique automaticamente: a. Atividade de voz/silêncio b. Os segmentos surdos/sonoros

```python
detecVoz = []
detecSilencio = []

for l in range(1, nFrames):
    
    ehVoz = 0
    ehSilencio = 0
    
    if ZCCs[l] > 15:
       
        ehVoz = 1
        ehSilencio = 0
        
    elif ZCCs[l] > 10:
       
        ehVoz = 0.5
        ehSilencio = 0
    
    else:
       
        ehSilencio = 1
        ehVoz = 0
        
    detecVoz.append(ehVoz)
    detecSilencio.append(ehSilencio)
```
